```python
df['FamilySize'] = df['SibSp'] + df['Parch']
df.drop(['SibSp','Parch'],axis=1 ,inplace=True)
df.corr()
```

```python
df.fillna(df.mean(), inplace=True)
df.isnull().sum()
```

```python
df.corr()
```

```python
df['Alone']= [0 if df['FamilySize'][i]>0 else 1 for i in df.index]
df.head()
```

```python
df.groupby(['Alone'])['Survived'].mean()
```

```python
df[['Alone', 'Fare']].corr()
```

```python
df['Sex']= [0 if df['Sex'][i]=='male' else 1 for i in df.index]
df.groupby(['Sex'])['Survived'].mean()
```

##**Conclusion**

1. The Fare charges increased with increase in Pclass.
2. The survival rate of high Pclass passengers were higher.
3. The survival rate of female was more than the male.
4. The survival rate of Passengers with family members aboard was higher than the Passengers boarding Alone.
5. The Passengers who boarded the ship at Cherbourg had better survival rate than other boarding ports , as the Cherbourg Port had maximum Passenger with 1st class seat and higher no. of Female passengers.  
6. The survival rate for Passengers between '20 to 50' age was hightest and lowest for passenger having age above 50 .
